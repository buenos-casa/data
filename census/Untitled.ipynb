```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
tmp = pd.read_json("data.json")
# Drop type label
dat = pd.read_json(tmp.properties.to_json())
dat = dat.T
```

```python
def cleanData(df):

    for col in df.columns:
        if '_q' in col or col == 'Comune' or col == 'RADPAIS_ID':
            df[col] = df[col].astype(int)
    
    #First, we define the column names used in all dataframes and their corresponding unified name
    columnNames = {
        'P19_q':'Education Quantile', 'P19_p':'Education Percent',
        'P05_q':'Immigration Quantile', 'P05_p':'Immigration Percent',
        'H2819C_q':'Cellular Quantile', 'H2819C_p':'Cellular Percent',
        'H2819B_q': 'Computer Quantile', 'H2819B_p': 'Computer Percent',
        'V02_q': 'Uninhabited Quantile', 'V02_p': 'Uninhabited Percent',
        'PROP_q':'Owner Quantile', 'PROP_p':'Owner Percent',
        'INQ_q': 'Rent Quantile', 'INQ_p': 'Rent Percent',
        'TENREG_q': 'Regular Quantile', 'TENREG_p': 'Regular Percent',
        'Comune': 'Commune', 'RADPAIS_ID': 'id'
    }

    #Then, delete empty rows, rename the columns and set the index for the Location (Name of the Barrio)
    df = (df.rename(columns=columnNames)
            .set_index('id'))
    
    df = df.drop(['REDCODE', 'DEPTO', 'TIPO', 'RADPAIS_', 'LINK', 'RADIO', 'PROV', 'PERIMETER','FRAC', 'AREA'], axis=1)
    
    return df
```

```python
dat = cleanData(dat)
```
